```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ELU(x, alpha=1.0):
    if x > 0:
        return x
    else:
        return alpha*(math.exp(x) - 1)

x = np.linspace(-10,6,10)
output = [ELU(i) for i in x]
plt.plot(x, output)
```

```python
def Hardshrink(x, lamba=0.5):
    if x > lamba:
        return x
    elif x < -lamba:
        return x
    else:
        return 0
    
x = np.linspace(-10,6,10)
output = [Hardshrink(i) for i in x]
plt.plot(x, output)
```

```python
def Hardsigmoid(x):
    if x <= -3:
        return 0
    elif x >= 3:
        return 1
    else:
        return x/6 + 1/2
    
x = np.linspace(-10,6,10)
output = [Hardsigmoid(i) for i in x]
plt.plot(x, output)
```

```python
def Hardtanh(x, min_val = -1.0, max_val = 1.0, min_value=None, max_value=None):
    if x > max_val:
        return max_val
    elif x < min_val:
        return min_val
    else:
        return x

x = np.linspace(-10,6,10)
output = [Hardtanh(i) for i in x]
plt.plot(x, output)
```

```python
def Hardswish(x):
    if x <= -3:
        return 0
    elif x >= 3:
        return x
    else:
        return x*(x+3)/6
    
x = np.linspace(-10,6,10)
output = [Hardswish(i) for i in x]
plt.plot(x, output)
```

```python
def LeakyReLU(x, negative_slope=0.01):
    if x >= 0:
        return x
    else:
        return negative_slope*x

x = np.linspace(-10,6,10)
output = [LeakyReLU(i) for i in x]
plt.plot(x, output)
```

```python
def LogSigmoid(x):
    return math.log(1/(1 + math.exp(-x)))

x = np.linspace(-10,6,10)
output = [LogSigmoid(i) for i in x]
plt.plot(x, output)
```

```python
def PReLU(x, num_parameters=1, init=0.25, device=None, dtype=None):
    if x >= 0:
        return x
    else:
        return x*init

x = np.linspace(-10,6,10)
output = [PReLU(i) for i in x]
plt.plot(x, output)
```

```python
def RELU(x):
    return max(0, x)

x = np.linspace(-10,6,10)
output = [RELU(i) for i in x]
plt.plot(x, output)
```

```python
def RELU6(x):
    return min(max(0, x), 6)

x = np.linspace(-2,9,10)
output = [RELU6(i) for i in x]
plt.plot(x, output)
```

```python
def SELU(x):
    scale = 1.0507009873554804934193349852946
    alpha = 1.6732632423543772848170429916717
    
    return scale*(max(0, x) + min(0, alpha*(math.exp(x) - 1)))

x = np.linspace(-10,6,10)
output = [SELU(i) for i in x]
plt.plot(x, output)
```
